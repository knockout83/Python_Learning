```python
# Subset UK data
uk = vax_data[['country','date','daily_vaccinations','total_vaccinations']][vax_data['country'].isin(['United Kingdom'])]
```

```python
# Sort UK data by date
uk.sort_values(by='date')
```

```python
# Assign initialisation for daily vaccinations from the first value in total vaccinations.
uk['daily_vaccinations2'] = uk['daily_vaccinations']
uk['daily_vaccinations2'].iloc[0] = uk['total_vaccinations'].iloc[0]
```

```python
# Calculate new cumulative figure for total vaccinations
uk['total_vaccinations2'] = uk['daily_vaccinations2'].cumsum()
```

```python
# View UK data.
# There's a slight discrepancy between the original and calculated total vaccinations figures.
uk
```

```python
# Melt the UK dataframe so we can compare the original and calculated figures.
ukmelt_daily = pd.melt(uk[['country', 'date', 'daily_vaccinations', 'daily_vaccinations2']], id_vars = ["country", 'date'], var_name="source", value_name="daily_vaccinations")
```

```python
ukmelt_total = pd.melt(uk[['country', 'date', 'total_vaccinations', 'total_vaccinations2']], id_vars = ["country", 'date'], var_name="source", value_name="total_vaccinations")
```

```python
# Plot bar chart of daily vaccinations
g = sns.catplot(
    data=ukmelt_daily, kind="bar",
    x="date", y="daily_vaccinations", hue="source",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("Date", "Daily Vaccinations")
g.legend.set_title("")
```

```python
# Plot bar chart of total vaccinations
g = sns.catplot(
    data=ukmelt_total, kind="bar",
    x="date", y="total_vaccinations", hue="source",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("Date", "Total Vaccinations")
g.legend.set_title("")
g.set_xticklabels(rotation=90)
```

```python
# Calculate difference between original and calculated figures
uk['daily_difference'] = uk['daily_vaccinations'] - uk['daily_vaccinations2']
uk['total_difference'] = uk['total_vaccinations'] - uk['total_vaccinations2']
```
